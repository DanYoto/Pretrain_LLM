```python
from datasketch import MinHash, MinHashLSH
from nltk import ngrams

data = [
    "minhash is a probabilistic data structure for estimating the similarity between datasets",
    "finhash dis fa frobabilistic dfata sftructure for festimating tfhe fimilarity fetween fatasets",
    "weights control the importance of each feature in the model",
    "wfrights cfontrol tfhe impfortance of each feafure in tfhe mfodel",
]

lsh = MinHashLSH(threshold=0.3, num_perm=128)
minhashes = {}
for c, i in enumerate(data):
    minhash = MinHash(num_perm=128)
    for d in ngrams(i, 3):
        minhash.update("".join(d).encode('utf-8'))
    lsh.insert(c, minhash)
    minhashes[c] = minhash

for i in range(len(minhashes.keys())):
    result = lsh.query(minhashes[i])
    print(result)
        
```

```python
# First step is to de-duplicate
# consider Minhash
from datasketch import MinHash, MinHashLSH
import re

NON_ALPHA = re.compile(r'[^a-zA-Z0-9]')

def get_min_hash(tokens, min_tokens = 128):
    if len(tokens) < min_tokens:
        return None
    min_hash = MinHash()
    for token in set(tokens):
        min_hash.update(token.encode())
    return min_hash

def get_token(inputs):
    return {token for token in NON_ALPHA.split(inputs) if len(token.strip()) > 0}
```
